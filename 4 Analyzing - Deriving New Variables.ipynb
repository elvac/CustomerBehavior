```python
# Import "orders_products_merged.pkl"
path = r'/Users/elvachen/Documents/DA Learning/Immersion-data resources'
df_ods_prds_merged = pd.read_pickle(os.path.join(path, 'prepared data', 'orders_products_merged.pkl'))
df_ods_prds_merged.head(100)
```

```python
# Create a subset dataframe with first one million rows
df_sub=df_ods_prds_merged[:1000000]
```

```python
# Define a user-defined function to flag prices
def price_label(row):

  if row['prices'] <= 5:
    return 'Low-range product'
  elif (row['prices'] > 5) and (row['prices'] <= 15):
    return 'Mid-range product'
  elif row['prices'] > 15:
    return 'High range'
  else: return 'Not enough data'
```

```python
# Create the new column by using the user-defined function
df_sub['price_range'] = df_sub.apply(price_label, axis=1)
```

```python
# Check the outputs
df_sub['price_range'].value_counts(dropna=False)
```

```python
# Check the maximum of the price
df_sub['prices'].max()
```

```python
# If-statements with loc() function
df_sub.loc[df_sub['prices'] > 15, 'price_range_loc'] = 'High-range product'
```

```python
df_sub.loc[(df_sub['prices'] <= 15) & (df_sub['prices'] > 5), 'price_range_loc'] = 'Mid-range product' 
```

```python
df_sub.loc[df_sub['prices'] <= 5, 'price_range_loc'] = 'Low-range product'
```

```python
df_sub['price_range_loc'].value_counts(dropna= False)
```

```python
# Set the busy-tier label by using for-loops
result = []

for value in df_ods_prds_merged["orders_day_of_week"]:
  if value == 0:
    result.append("Busiest day")
  elif value == 4:
    result.append("Least busy")
  else:
    result.append("Regularly busy")
```

```python
# 4.7.2 Create a new column to label two busiest days and two slowest days
busy_tiers=[]

for value in df_ods_prds_merged['orders_day_of_week']:
    if value==0 :
        busy_tiers.append('First busiest day')
    elif value==1:
        busy_tiers.append('Second busiest day')
    elif value==4:
        busy_tiers.append('First lowest day')
    elif value==3:
        busy_tiers.append('Second lowest day')
    else:
        busy_tiers.append('Regular day')
```
