```python
# 4.8.6 Create a spending flag for each user based on the average price across all their orders using specific criteria
    # Create a new column to contain average price for each customer
df_ods_prds['avg_price']=df_ods_prds.groupby('user_id')['prices'].transform(np.mean)
df_ods_prds.head(30)
```

```python
    # Create a new column to conctain spending flag
df_ods_prds.loc[df_ods_prds['avg_price'] < 10, 'spending_flag'] = 'Low spender'
```

```python
df_ods_prds.loc[df_ods_prds['avg_price'] >= 10, 'spending_flag'] = 'High spender'
```

```python
# Check the output
df_ods_prds['spending_flag'].value_counts(dropna= False)
```

```python
# 4.8.7 Create an order frequency flag that marks the regularity of a user’s ordering behavior according to the median in the “days_since_prior_order” column
    # Create a new column to contain median of 'days_since_prior_order'
df_ods_prds['median_days_s_p_ods']=df_ods_prds.groupby('user_id')['days_since_prior_order'].transform(np.median)
df_ods_prds.head(30)
```

```python
    # Create a new column to contain frequency flag
fqcy_flag=[]

    for value in df_ods_prds['median_days_s_p_ods']:
        if value > 20:
            fqcy_flag.append('Non-frequent customer')
        elif (value > 10) & (value <= 20):
            fqcy_flag.append('Regular customer')
        else:
            fqcy_flag.append('Frequent customer')
df_ods_prds['frequency_flag']= fqcy_flag
df_ods_prds.head()
```

```python
# Check to output
df_ods_prds['frequency_flag'].value_counts(dropna= False)
```
